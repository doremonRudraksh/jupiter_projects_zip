```python
from IPython.display import Image
import pandas as pd
from matplotlib import pyplot as plt

df =  pd.read_csv("Titanic.csv")
df
df = df.dropna()
df
passenger_survived = df.loc[df["Survived"]==1]
passenger_survived

plt.figure(figsize = (45,45))
plt.hist(passenger_survived["Age"], bins=10)
plt.title("No. of people survived as per age group")
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

```python
passenger_died = df.loc[df["Survived"] == 0]
passenger_died

plt.figure(figsize = (7,7))
plt.hist(passenger_died["Age"],bins=10)
plt.title("No. of Deaths as per age groups")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

```

```python
Image(filename = "titanic_classes.png")

total_classes_1 = df.loc[df["Pclass"] ==1]["Pclass"].count()
total_classes_1

survived = df.loc[(df["Survived"] == 1) & (df["Pclass"]==1)]
survived_1 = survived["Pclass"].count()
survived_1

total_classes_2 = df.loc[df["Pclass"] ==2]["Pclass"].count()
total_classes_2

survived2 =  df.loc[(df["Survived"] ==1)& (df["Pclass"] ==2)]
survived_2 = survived2["Pclass"].count()
survived_2

total_classes_3 = df.loc[df["Pclass"] ==3]["Pclass"].count()
total_classes_3

survived3 = df.loc[(df["Survived"] ==1) & (df["Pclass"]==3)]
survived_3 = survived3["Pclass"].count()
survived_3

total_passengers = [total_classes_1,total_classes_2,total_classes_3]
total_survived = [survived_1,survived_2,survived_3]

index = ["total-vs-survived in class 1","total-vs-survived in class 2","total-vs-survived in class 3"]
plt.bar(index,total_passengers , bottom = total_survived,color="lightskyblue",label = "Total People")
plt.bar(index,total_survived,color="red",label="Survived People")
plt.xticks(rotation = 45)
plt.legend()
plt.show()
```
