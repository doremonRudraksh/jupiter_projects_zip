```python
#Group by age column and find average heart rate at different age
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("framingham.csv")
df

non_smokers = df.loc[df["currentSmoker"]==0]
non_smokers

group_non_smokers = non_smokers.groupby("age")["heartRate"].mean().reset_index()
group_non_smokers

smokers = df.loc[df["currentSmoker"] ==1]
smokers

group_smoker = smokers.groupby("age")["heartRate"].mean().reset_index()
group_smoker

plt.figure(figsize = (19,8))

plt.plot(group_non_smokers["age"],group_non_smokers["heartRate"],label = "HEARTRATE OF NON SMOKERS" , linewidth=3.0)

plt.plot(group_smoker["age"],group_smoker["heartRate"],label="HEARTRATE OF SMOKERS",linewidth=3.0)

plt.xlabel("age")
plt.xticks(rotation="vertical")
plt.ylabel("Heart rate of a smoker and a non smoker")
plt.title("Comparision between the heartrate of a smokers and a non-smoker")
plt.legend()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
df_1 = pd.read_csv("framingham.csv")
df_1

coronary_heart_diseases = df_1.loc[df["TenYearCHD"] ==1]
coronary_heart_diseases

group_age = coronary_heart_diseases.groupby("age")["TenYearCHD"].mean().reset_index()
group_age

plt.figure(figsize = (12,8))
rgb = np.random.rand(3)
print(rgb)
plt.scatter(group_age["age"],group_age["TenYearCHD"],c=[rgb],label="Chance of CHD in next ten years",alpha=0.5,s=65)
plt.xlabel("age")
plt.xticks(rotation="vertical")
plt.ylabel("chance of chd in next ten years")
plt.title("Chance of CHD in next ten years as per age")
plt.legend()
plt.show()
```
