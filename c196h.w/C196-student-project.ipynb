# Activity - 1 Clean the data and show which top 10 countries has the highest Undernourished rate as compared with their Population

```python
#import the required packages 
import pandas as pd
import matplotlib.pyplot as plt

#Read the csv file.
df = pd.read_csv("COVID-19 Healthy Diet Dataset.csv")
df

df.replace("-",value="NaN", inplace=True)
df.dropna()
df.replace("<2.5",int(3),inplace=True)
df

df["Undernourished"] = df["Undernourished"].astype(float)
df

group_country = df.groupby("Country")["Undernourished","Population"].sum().reset_index()
group_country

sorted_by_unnourished = group_country.sort_values(by="Undernourished", ascending = False)
sorted_by_unnourished

top10 = sorted_by_unnourished.head(10)
top10["percentage_number"] = top10["Undernourished"]/100* top10["Population"]
top10

total_population = top10['Population']
percent_number = top10['percentage_number']
total_countries = top10['Country']
plt.figure(figsize = (10, 10))
plt.title("Population vs Undernourished Rate")
plt.bar(total_countries, total_population, color = 'lightskyblue', label = 'Country Population')
plt.bar(total_countries, percent_number, color = 'red', label = 'Undernourished')
plt.xticks(rotation = 80)
plt.legend()
plt.xlabel("Countries")
plt.ylabel("Population")
plt.show()

```
