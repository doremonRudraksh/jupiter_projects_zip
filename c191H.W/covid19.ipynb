```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
sorted_df = df.sort_values(by=["total_cases"])
sorted_df
lowest_cases = sorted_df["total_cases"].head(5)
lowest_country = sorted_df["country"].head(5)
lowest_cases
lowest_country
plt.ylabel("country")
plt.xlabel("Number of cases")
plt.xticks(rotation = "vertical")
label= lowest_cases
name = lowest_country
plt.bar(name,label,width=0.4,color = ("red","green","blue","yellow","pink"))
```

```python
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
sorted_deaths = df.sort_values(by=["total_deaths"])
sorted_deaths
highest_deaths = sorted_deaths["total_deaths"].tail(5)
highest_death_country = sorted_deaths["country"].tail(5)
highest_deaths
highest_death_country
plt.ylabel("country")
plt.xlabel("Number of deaths")
plt.xticks(rotation = "vertical")
label= highest_deaths
name = highest_death_country
plt.bar(name,label,width=0.4,color = ("red","green","blue","yellow","pink"))
```

```python
#Task 5
#Sort the data as per active cases
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
sorted_deaths = df.sort_values(by=["active_cases"])
sorted_deaths
highest_active_cases= sorted_deaths["active_cases"].tail(5)
highest_active_cases_country= sorted_deaths["country"].tail(5)
highest_active_cases
highest_active_cases_country
plt.ylabel("country")
plt.xlabel("Number of Active cases")
plt.xticks(rotation = "vertical")
label= highest_active_cases
name = highest_active_cases_country
plt.bar(name,label,width=0.4,color = ("red","green","blue","yellow","pink"))
```
