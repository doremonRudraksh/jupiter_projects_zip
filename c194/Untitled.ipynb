```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

disney_movies = pd.read_csv("disney_movies.csv")
disney_movies

movies_collection = disney_movies.loc[disney_movies["total_gross"]>400000000]
movies_collection

plt.figure(figsize=(12,8))
plt.xticks(rotation = "vertical")
rgb = np.random.rand(3)
print(rgb)

plt.scatter(movies_collection["movie_title"],movies_collection["total_gross"],c=[rgb],alpha=0.5,s=movies_collection["total_gross"]/1000000)
plt.legend()
plt.xlabel("movies",size=14)
plt.ylabel("total gross", size=14)
plt.show()




```
