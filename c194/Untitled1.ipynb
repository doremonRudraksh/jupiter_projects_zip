```python
from matplotlib import pyplot as plt
import pandas as pd

movies_update = pd.read_csv("movie_update.csv")
movies_update

group_year = movies_update.groupby("Year")["Title"].count().reset_index()
group_year

plt.figure(figsize=(20,8))

plt.scatter(group_year["Year"], group_year["Title"],c = "red", label = "No. of movies", alpha=0.5, s= group_year["Title"]*3)

plt.xticks(rotation = "vertical")
plt.legend()
plt.xlabel("Years",size=14)
plt.ylabel("No. of movies",size=14)
plt.show()


```
