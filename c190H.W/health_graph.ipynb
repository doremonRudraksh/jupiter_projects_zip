```python
#import the libraries
import pandas as pd
from matplotlib import pyplot as plt

#Task 1
#Read the bmi.csv

dataframe = pd.read_csv("bmi.csv")
dataframe
bmi_top5 = dataframe.head(5)
name = bmi_top5["age"]
number = bmi_top5["bmi"]

plt.ylabel("BMI")
plt.xlabel("Age")
plt.xticks(rotation = "vertical")

plt.bar(name,number,width=0.4,color = ("red","blue","green","yellow","purple"))
```

```python
bp_df = pd.read_csv('blood_pressure.csv')
print(bp_df)
bp_top_5 = bp_df.tail(5)
name = bp_top_5['Age']
number = bp_top_5['BMI']

plt.ylabel("Blood Pressure")
plt.xlabel("Age")
plt.xticks(rotation = 'vertical')

plt.bar(name, number, width = 0.4, color = ('red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet'))
```
